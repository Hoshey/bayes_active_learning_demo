### fuction for generate data
#### size=300, dim=2

```python
# k --> sample size
# n --> variable size
def dta(k, n) :
    rd.seed(3)
    state = [[1, 100], [101, 200], [301, 400], [201, 300],
             [1, 10], [11, 15], [16, 20], [21, 25],
             [10, 100], [101, 150], [151, 200], [201, 250],
             [6, 18], [21, 50], [51, 100], [101, 120],
             [2, 20], [30, 90], [1, 5], [6, 10]]
    y1= [rd.randint(1, 1) for _ in range( int(k/2) )]
    y0= [rd.randint(0, 0) for _ in range( int(k/2) )]
    print(len(y1))
    print(len(y0))

    y1.extend(y0)
    print(len(y1))
    print('-'*30)
    
    r = []
    j = 1

    for i in range(n) :    
        r0 = [rd.randint(state[j-1][0], state[j-1][1]) for _ in range( int(k/2) )]
        r1 = [rd.randint(state[j][0], state[j][1]) for _ in range( int(k/2) )] 
        print(len(r0))
        print(len(r1))

        r0.extend(r1)
        print(len(r0))
        j += 2

        r.append(r0)
        print(len(r))
        print('_'*30)
        
    ra = np.transpose( np.asarray(r) )
    ra.shape

    np.random.seed(3)
    np.random.shuffle(ra)
    print(ra[0:9])
    np.random.seed(3)
    np.random.shuffle(y1)
    print(y1[0:9])
    print('='*60)

    return ra , y1

```

```python
x, y = dta(300, 2)

plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.axis(lim)
```
